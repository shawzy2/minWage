```python
def parse_team_name(column):
    return column.find('h6').find('div').get_text().split(' Acquire:')[0]

def parse_assets(column):
    assets = []
    for div in column.find_all('div')[2].find_all('div'):
        if len(div.get_text()) == 0:
            continue

        asset_link = ''
        asset_name = div.get_text()
        if div.find('a'):
            asset_link = div.find('a').get('href')

        assets.append([asset_name, asset_link])
    return assets
```
